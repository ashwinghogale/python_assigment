```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
epl  = pd.read_csv('EPL_20_21.csv')
```

```python
epl.head()
```

```python
epl.info()
```

```python
 epl.isnull().sum()
```

```python
##create new colums
epl['minpermatch'] = epl['Mins']/epl['Matches'].astype(int)
epl['goalpermatch'] = epl['Goals']/epl['Matches'].astype(float)
```

```python
epl.head()
```

```python
##total goals 
a = epl[['Club','Goals']].groupby(epl['Club']).sum()
a.plot(kind='bar',color='red')

```

```python
## penalty attempt
b = epl[['Penalty_Attempted','Club']].groupby(epl['Club']).sum()
b.sort_values(by='Penalty_Attempted',ascending=True,inplace=True)
b.plot(kind='barh',color='blue')
```
