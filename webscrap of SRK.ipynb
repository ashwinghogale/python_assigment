```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

```python
#get the url 
url = 'https://www.bollywoodhungama.com/celebrity/shah-rukh-khan/box-office/#bh-micro-tab-content'
## fetch the data
response=requests.get(url)
```

```python
response.content
```

# Step 2

```python
##put the data in beautyful soup
soup = BeautifulSoup(response.content,'html.parser')
soup
```

```python
## get the tabledata 
data = soup.find('table',{'id':'example'})
dat = data.findAll('span')
```

# step 3 

```python
## extract only text data from span
d = [x.text for x in dat]
d.pop()
d = d[0:-23]
d.append(0)
d.append(0)
d.append(0)
```

```python
da = np.array(d).reshape(59,5)
```

```python
df = pd.DataFrame(da)
df
```

```python
##give column name
col = ['date','open','weekend','week','final']
df =pd.DataFrame(da,columns=col)
df
```
