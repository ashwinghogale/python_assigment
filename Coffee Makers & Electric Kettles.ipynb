```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
url ='https://www.snapdeal.com/products/appliances-coffee-makers-kettles?sort=plrty&q=avgRating%3A4.0%7C'
html = requests.get(url)
html.content
```

```python
soup = BeautifulSoup(html.content,'html.parser')
soup
```

```python
data =soup.find('div',{'id':'products'})
author_title=data.findAll('p')
##<span> for price,discount,percentoff <p> for authorname,title
```

```python
author_title
```

```python
author=[x.text for x in author_title ]
author
```

```python
a =np.array(author)
a.size
a_name=a.reshape(20,2)
```

```python
df = pd.DataFrame(a_name)
df
```

# for price and ratings


```python
pr= data.findAll('span')
```

```python
pri = [x.text for x in pr]
pri
```

```python
pri=np.array(pri)
pric=pri.reshape(20,4)
```

```python
price = pd.DataFrame(pric)
price
```

# combine two data frame

```python
cols = {0:'name',1:'No_review'}
df = df.rename(columns=cols)
df
```

```python
price_cols = {1:'MRP',2:'Discounted Price',3:'Discount'}
price=price.rename(columns=price_cols) 
```

```python
price_df=price.drop(columns=[0])
price_df
```

# Combine two Dataframe

```python
CoffeeMaker=df.join(price_df)
```

```python
CoffeeMaker.to_csv('D://Coffeemakere.csv',index=False)
```
